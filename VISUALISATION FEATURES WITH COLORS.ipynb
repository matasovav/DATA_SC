```python

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
sc_min = pd.read_csv('sc_min.csv', header=0, sep=',')
sc_max = pd.read_csv('sc_max.csv', header=0, sep=',')
sc_mean = pd.read_csv('sc_mean.csv', header=0, sep=',')
```

```python
df = sc_min
df.sort_values('critical_temp', inplace=True)
df
```

```python
df = pd.read_csv('train.csv')
els = pd.read_csv('unique_m.csv')
df = pd.concat([df, els], axis=1)
df = df.loc[:,~df.columns.duplicated()]
df.sort_values('critical_temp', inplace=True)
df
```

```python
# Get features


features = [col.split('_', 1)[-1] for col in df.columns if col.startswith('mean_')]
features = tuple(features)
features
```

```python
# Normalize features

for col in df.columns:
    if not col.endswith(features):
        continue
    vals = df[col]
    df[col] = np.interp(vals, (vals.min(), vals.max()), (-1, +1))
```

```python
# Plot feature distribution

fig, axes = plt.subplots(2, 4, figsize=(50, 50))
fig.tight_layout()

for i, ax in enumerate(axes.flatten()):
    vals = np.vstack(df[[col for col in df.columns if col.endswith(features[i])]].values)
    ax.set_title(features[i])
    ax.imshow(vals, origin='upper', extent=[1,10, 1,10])
```
